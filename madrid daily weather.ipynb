### 6. Analysis with NumPy & Pandas

```python
# Remove leading/trailing spaces from all column names
df.columns = df.columns.str.strip()

import numpy as np

# Average temperatures
avg_max_temp = np.mean(df['Max TemperatureC'])
avg_mean_temp = np.mean(df['Mean TemperatureC'])
avg_min_temp = np.mean(df['Min TemperatureC'])
print("Average Max Temperature (Â°C):", avg_max_temp)
print("Average Mean Temperature (Â°C):", avg_mean_temp)
print("Average Min Temperature (Â°C):", avg_min_temp)

# Average humidity
avg_max_humidity = np.mean(df['Max Humidity'])
avg_mean_humidity = np.mean(df['Mean Humidity'])
avg_min_humidity = np.mean(df['Min Humidity'])
print("Average Max Humidity (%):", avg_max_humidity)
print("Average Mean Humidity (%):", avg_mean_humidity)
print("Average Min Humidity (%):", avg_min_humidity)

# Average wind speed
avg_wind_speed = np.mean(df['Mean Wind SpeedKm/h'])
print("Average Wind Speed (Km/h):", avg_wind_speed)
```

### Insights:
 
1.The average Max Temperature is around XÂ°C and the average Min Temperature is around YÂ°C, showing that Madrid experiences moderate daily temperatures.

2.Average humidity is about Z%, indicating a generally moderate climate with occasional very dry or very humid days.

3.The average wind speed is around W Km/h, suggesting mild to moderate wind conditions on most days.

### 7. Visualization with Matplotlib

```python
# Ensure CET is datetime
df['CET'] = pd.to_datetime(df['CET'])

# Extract month
df['Month'] = df['CET'].dt.month

# Average Max Temperature by Month
avg_max_temp_by_month = df.groupby('Month')['Max TemperatureC'].mean()

# Bar chart
months = avg_max_temp_by_month.index
temps = avg_max_temp_by_month.values

plt.bar(months, temps, color='skyblue')
plt.title("Average Max Temperature by Month")
plt.xlabel("Month")
plt.ylabel("Average Max Temperature (Â°C)")
plt.show()
```

### Insights

1.The highest average temperatures occur in July and August, reflecting the peak of summer in Madrid.

2.The lowest average temperatures occur in January and February, showing the cold winter months.

```python
df['CET'] = pd.to_datetime(df['CET'])

# Extract month
df['Month'] = df['CET'].dt.month

# Average Max Temperature by Month
avg_max_temp_by_month = df.groupby('Month')['Max TemperatureC'].mean()

# Line plot
plt.plot(avg_max_temp_by_month.index, avg_max_temp_by_month.values, marker='o', color='orange')
plt.title("Average Max Temperature by Month")
plt.xlabel("Month")
plt.ylabel("Average Max Temperature (Â°C)")
plt.xticks(range(1,13))  # Show all months on x-axis
plt.show()
```

### Insights

1.Temperatures gradually increase from January to July and then decrease towards December, forming a clear seasonal pattern.

2.The warmest months are July and August, while January and December are the coldest, indicating typical Mediterranean climate trends.

### 8. Visualization with Seaborn

```python
sns.histplot(df['Max TemperatureC'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Max Temperature (Â°C)")
plt.xlabel("Max Temperature (Â°C)")
plt.ylabel("Frequency")
plt.show()
```

### Insights

1.Most days in Madrid have moderate temperatures, with a peak around 15â€“25Â°C, reflecting mild weather for most of the year.

2.Very few days are extremely cold (<0Â°C) or extremely hot (>35Â°C), showing that extreme temperatures are rare.

```python
# Create Temperature Category
df['Temp Category'] = df['Max TemperatureC'].apply(lambda x: 'Hot (>=25Â°C)' if x >= 25 else 'Cold (<25Â°C)')

# Extract Month
df['Month'] = df['CET'].dt.month

# Countplot of Temperature Category by Month
plt.figure(figsize=(12,6))
sns.countplot(x='Month', hue='Temp Category', data=df, palette='coolwarm')
plt.title("Count of Hot and Cold Days by Month")
plt.xlabel("Month")
plt.ylabel("Number of Days")
plt.show()
```

### Insights

1.OThe number of Hot days increases during summer months (Juneâ€“August) and decreases during winter months (Decemberâ€“February).

2.Most winter months have predominantly Cold days, showing the seasonal effect on Madridâ€™s daily maximum temperatures.

```python
df['Month'] = df['CET'].dt.month

# Boxplot of Max Temperature by Month
plt.figure(figsize=(12,6))
sns.boxplot(x='Month', y='Max TemperatureC', data=df, palette='coolwarm')
plt.title("Max Temperature Distribution by Month")
plt.xlabel("Month")
plt.ylabel("Max Temperature (Â°C)")
plt.show()
```

### Insights

1.Summer months (Juneâ€“August) have higher median temperatures and a wider spread, showing hot days and occasional extreme highs.

2.Winter months (Decemberâ€“February) have lower median temperatures with fewer extreme highs, indicating a cooler and more stable temperature range.

```python
df['Temp Category'] = df['Max TemperatureC'].apply(lambda x: 'Hot (>=25Â°C)' if x >= 25 else 'Cold (<25Â°C)')

# Select numeric features for pairplot
numeric_features = ['Max TemperatureC', 'Mean TemperatureC', 'Min TemperatureC', 'Mean Humidity']

# Pairplot with Temp Category as hue
sns.pairplot(df[numeric_features + ['Temp Category']], hue='Temp Category', palette='coolwarm')
plt.show()
```

### Insights

1.Hot days usually have higher max, mean, and min temperatures, and the points for Hot and Cold days are clearly separated.

2.There may be a negative correlation between temperature and humidity, showing that hotter days tend to have lower mean humidity.

```python
numeric_features = ['Max TemperatureC', 'Mean TemperatureC', 'Min TemperatureC', 
                    'Mean Humidity', 'Max Wind SpeedKm/h']

# Compute correlation
corr = df[numeric_features].corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Weather Features")
plt.show()
```

### Insights

1.Max, Mean, and Min Temperatures are strongly positively correlated, as expected.

2.Temperature and Mean Humidity show a slight negative correlation, suggesting hotter days tend to be less humid.

3.Wind speed appears largely independent of temperature and humidity, with low correlation values.
