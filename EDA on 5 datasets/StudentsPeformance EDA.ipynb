```python
# 1.Load libraries & Dataset
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
# 2. Basic Information & Overview
print("Shape of the dataset:",df.shape)
print("\nData Info:\n")
df.info()

# Summary Statistics
df.describe(include='all')
```

```python
# 3.Missing values analysis 
missing=pd.DataFrame({
    'missing_count':df.isnull().sum(),
    'missing_percent':(df.isnull().sum()/len(df))*100
})
missing.sort_values(by='missing_percent',ascending=False)
```

```python
# 3.Missing values analysis after cleaning
missing=pd.DataFrame({
    'missing_count':df.isnull().sum(),
    'missing_percent':(df.isnull().sum()/len(df))*100
})
missing.sort_values(by='missing_percent',ascending=False)
```

#### Insight: All columns in the dataset are complete with no missing values. Therefore, no imputation or removal was required, ensuring that the dataset retains its full informational value for further analysis.

```python
# Duplicates & Data type check

# Check duplicate rows
print("Duplicate rows:", df.duplicated().sum())

# Convert suitable columns to categorical type
df['Sex'] = df['Sex'].astype('category')
df['Graduated high-school type'] = df['Graduated high-school type'].astype('category')
df['Scholarship type'] = df['Scholarship type'].astype('category')
df['Additional work'] = df['Additional work'].astype('category')
df['Regular artistic or sports activity'] = df['Regular artistic or sports activity'].astype('category')
df['Do you have a partner'] = df['Do you have a partner'].astype('category')
df['COURSE ID'] = df['COURSE ID'].astype('category')
```

```python
df.info()
```

```python
# 6. Univariate Analysis - Numerical Features

# Selecting numeric columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
print(df[num_cols].describe())

# Histogram: Student Age
sns.histplot(df['Student Age'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Student Age')
plt.xlabel('Student Age')
plt.ylabel('Count')
plt.show()

# Boxplot: Cumulative GPA (last semester)
sns.boxplot(x=df['Cumulative grade point average in the last semester (/4.00)'], color='orange')
plt.title('Distribution of Cumulative GPA (Last Semester) with Outliers')
plt.xlabel('Cumulative GPA (Last Semester)')
plt.show()
```

```python
# 7. Univariate Analysis - Categorical Features

cat_cols = ['Sex',
            'Graduated high-school type',
            'Scholarship type',
            'Additional work',
            'Regular artistic or sports activity',
            'Do you have a partner',
            'COURSE ID']

for col in cat_cols:
    sns.catplot(x=col, data=df, kind='count', palette='pastel')
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# 8. Bivariate Analysis - Numerical vs Categorical

# GPA (Last Semester) by Scholarship Type
sns.boxplot(x='Scholarship type', 
            y='Cumulative grade point average in the last semester (/4.00)', 
            data=df, palette='Set2')
plt.title("GPA (Last Semester) by Scholarship Type")
plt.xlabel("Scholarship Type")
plt.ylabel("Cumulative GPA (Last Semester)")
plt.show()

# Expected GPA (Graduation) by Gender
sns.boxplot(x='Sex', 
            y='Expected Cumulative grade point average in the graduation (/4.00)', 
            data=df, palette='Set3')
plt.title("Expected GPA (Graduation) by Gender")
plt.xlabel("Sex")
plt.ylabel("Expected GPA (Graduation)")
plt.show()
```

```python
# 10. Bivariate Analysis - Categorical vs Categorical

# Scholarship Type by Gender
sns.countplot(x='Scholarship type', hue='Sex', data=df, palette='pastel')
plt.title("Scholarship Type by Gender")
plt.xlabel("Scholarship Type")
plt.ylabel("Count")
plt.show()

# Relationship Status vs Sports/Artistic Activity
sns.countplot(x='Do you have a partner', hue='Regular artistic or sports activity', data=df, palette='Set2')
plt.title("Relationship Status vs Sports/Artistic Activity")
plt.xlabel("Do you have a partner")
plt.ylabel("Count")
plt.show()
```

```python
# 9. Correlation and Multivariate Analysis

# Select key numerical columns for correlation
corr = df[['Student Age',
           'Cumulative grade point average in the last semester (/4.00)',
           'Expected Cumulative grade point average in the graduation (/4.00)',
           'GRADE']].corr()

# Correlation heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Key Numerical Features")
plt.show()

# Pairplot to visualize relationships
sns.pairplot(df[['Student Age',
                 'Cumulative grade point average in the last semester (/4.00)',
                 'Expected Cumulative grade point average in the graduation (/4.00)',
                 'GRADE']])
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

### Insights: 

1.Cumulative GPA (last semester) has a strong positive correlation with expected graduation GPA, indicating that students who perform well currently are likely to maintain high performance until graduation.

2.Student Age shows little to no correlation with grades, suggesting that academic performance is not strongly influenced by age in this dataset.

3.RADE is moderately positively correlated with both last semester and expected GPA, meaning higher course grades reflect overall academic consistency.

```python
# 11. Outlier Detection using IQR for Cumulative GPA (Last Semester)

col = 'Cumulative grade point average in the last semester (/4.00)'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df[col] < lower) | (df[col] > upper)]
print(f"Number of outliers in {col}:", len(outliers))

# Cap outliers
df[col + '_capped'] = df[col].clip(lower, upper)

# Boxplot after capping
sns.boxplot(x=df[col + '_capped'], color='orange')
plt.title(f'{col} after outlier capping')
plt.xlabel(col)
plt.show()
```

```python
# 12. Feature Engineering

# Create new features
df['has_partner'] = (df['Do you have a partner'] == 2).astype(int)  # assuming 2 = Yes
df['engaged_in_activity'] = (df['Regular artistic or sports activity'] == 2).astype(int)  # assuming 2 = Yes

# Visualize
sns.countplot(x='has_partner', hue='engaged_in_activity', data=df, palette='pastel')
plt.title("Engagement in Activities by Partner Status")
plt.xlabel("Has Partner")
plt.ylabel("Count")
plt.show()

# Display first rows of new features
df[['has_partner', 'engaged_in_activity', 'Sex', 'Scholarship type']].head()
```

```python
# Export the Cleaned data
df.to_csv("studentperformance_cleaned.csv",index=False)
print("\n Cleaned StudentPerformance data saved as studentperformance_cleaned.csv ")
```
