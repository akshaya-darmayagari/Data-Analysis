```python
# 1.Load libraries & Dataset
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
# 2. Basic Information & Overview
print("Shape of the dataset:",df.shape)
print("\nData Info:\n")
df.info()

# Summary Statistics
df.describe(include='all')
```

```python
# 3. Missing Values Analysis

missing = pd.DataFrame({
    'missing_count': df.isnull().sum(),
    'missing_percent': (df.isnull().sum() / len(df)) * 100
})

missing.sort_values(by='missing_percent', ascending=False)
```

```python
# 4. Handling Missing Values

# Fill numeric missing values with the mean
df['Postal Code'].fillna(df['Postal Code'].mean(), inplace=True)

# Check missing values after cleaning
print("Missing values after cleaning:", df.isnull().sum().sum())
```

```python
# 3.Missing values analysis after cleaning
missing=pd.DataFrame({
    'missing_count':df.isnull().sum(),
    'missing_percent':(df.isnull().sum()/len(df))*100
})
missing.sort_values(by='missing_percent',ascending=False)
```

#### Insight: We filled important columns and removed highly missing values.This keeps maximum information while making 

```python
# 5. Duplicates & Data Type Check

# Check for duplicate rows
print("Duplicate rows: ", df.duplicated().sum())

# Convert relevant columns to category type
cat_cols = ['Segment', 'Ship Mode', 'Category', 'Sub-Category', 'Region']
for col in cat_cols:
    df[col] = df[col].astype('category')

# Check data types after conversion
print(df.dtypes)
```

```python
df.info()
```

```python
## 6. Univariate Analysis - Numerical Features

# Select numeric columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
print(df[num_cols].describe())

# Histogram: Sales
sns.histplot(df['Sales'], bins=30, kde=True, color='skyblue')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Count')
plt.show()

# Boxplot: Sales
sns.boxplot(x=df['Sales'], color='orange')
plt.title('Sales Distribution with Outliers')
plt.xlabel('Sales')
plt.show()
```

```python
# 7. Univariate Analysis - Categorical Features

cat_cols = df.select_dtypes('category').columns

for col in cat_cols:
    sns.countplot(x=col, data=df, palette='pastel')
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# 8. Bivariate Analysis - Numerical vs Categorical

# Sales by Segment
sns.boxplot(x='Segment', y='Sales', data=df, palette='Set2')
plt.title("Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.show()

# Sales by Category
sns.boxplot(x='Category', y='Sales', data=df, palette='Set3')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()

# Sales by Ship Mode
sns.boxplot(x='Ship Mode', y='Sales', data=df, palette='pastel')
plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Sales")
plt.show()
```

```python
# 9. Bivariate Analysis - Categorical vs Categorical

# Segment vs Ship Mode
sns.countplot(x='Segment', hue='Ship Mode', data=df, palette='Set2')
plt.title("Orders by Segment and Ship Mode")
plt.xlabel("Segment")
plt.ylabel("Order Count")
plt.show()

# Category vs Region
sns.countplot(x='Category', hue='Region', data=df, palette='Set3')
plt.title("Orders by Category and Region")
plt.xlabel("Category")
plt.ylabel("Order Count")
plt.show()

# Sub-Category vs Ship Mode
sns.countplot(x='Sub-Category', hue='Ship Mode', data=df, palette='pastel')
plt.title("Orders by Sub-Category and Ship Mode")
plt.xlabel("Sub-Category")
plt.ylabel("Order Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# 10. Correlation and Multivariate Analysis

# Correlation between numeric features
corr = df[['Sales', 'Postal Code', 'Row ID']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Pairplot for numeric features (focus on Sales)
sns.pairplot(df[['Sales', 'Postal Code', 'Row ID']])
plt.show()
```

### Insights: 

1.Sales shows little to no correlation with Postal Code or Row ID  as expected, location or row index doesnt strongly affect sales.

2.Sales distribution is highly skewed  most orders have relatively low sales, but a few very high-value orders exist.

3.High-value orders are potential outliers, which can impact mean sales if not considered separately.

```python
# 11. Outlier Detection using IQR for Sales

Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['Sales'] < lower) | (df['Sales'] > upper)]
print("Number of outliers in Sales:", len(outliers))

# Cap outliers
df['Sales_capped'] = df['Sales'].clip(lower, upper)

# Boxplot after capping
sns.boxplot(x=df['Sales_capped'], color='orange')
plt.title('Sales after Outlier Capping')
plt.xlabel('Sales')
plt.show()
```

```python
# 12. Feature Engineering

# Example 1: Extract Order Month and Year from Order Date
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Order_Year'] = df['Order Date'].dt.year
df['Order_Month'] = df['Order Date'].dt.month

# Example 2: Calculate shipping duration in days
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
df['Shipping_Duration'] = (df['Ship Date'] - df['Order Date']).dt.days

# Example 3: Flag high-value orders (Sales > 1000)
df['High_Value_Order'] = (df['Sales'] > 1000).astype(int)

# Visualize High Value Orders by Segment
sns.countplot(x='High_Value_Order', hue='Segment', data=df, palette='Set2')
plt.title("High Value Orders by Segment")
plt.xlabel("High Value Order (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()

# Display first rows
df[['Order ID', 'Sales', 'High_Value_Order', 'Segment', 'Shipping_Duration']].head()
```

```python
# Export the cleaned data as training dataset
df.to_csv("train_cleaned.csv", index=False)
print("\nCleaned training data saved as train_cleaned.csv")
```
