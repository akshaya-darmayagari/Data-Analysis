```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('loan_approved.csv')
mean=df['ApplicantIncome'].mean()
print('Mean',mean)
std=df['ApplicantIncome'].std()
print(std)
print(df['ApplicantIncome'].head(5))
z1=(5849-mean)/std
print(z1)
z2=(4582-mean)/std
print(z2)
z3=(3000-mean)/std
print(z3)
z4=(2583-mean)/std
print(z4)
z5=(6000-mean)/std
print(z5)
range=df['ApplicantIncome'].max()-df['ApplicantIncome'].min()
print(range)
```

```python
m1=df['ApplicantIncome'].skew()
k1=df['ApplicantIncome'].kurt()
print('ApplicantIncome\nSkewness:',m1,' Kurtosis:',k1)
m2=df['CoapplicantIncome'].skew()
k2=df['CoapplicantIncome'].kurt()
print('CoapplicantIncome\nSkewness:',m2,' Kurtosis:',k2)
m3=df['LoanAmount'].skew()
k3=df['LoanAmount'].kurt()
print('LoanAmount\nSkewness:',m3,' Kurtosis:',k3)
m4=df['Loan_Amount_Term'].skew()
k4=df['Loan_Amount_Term'].kurt()
print('Loan_Amount_Term\nSkewness:',m4,' Kurtosis:',k4)
```

ApplicantIncome->Highly skewed(right),Leptokurtic

CoapplicantIncome->Highly skewed(right),Leptokurtic

LoanAmount->Highly skewed(right),Leptokurtic

Loan_Amount_Term->Highly skewed(left),Leptokurtic

### One hot Encoding

```python
df['Property_Area'].unique()
```

```python
df['Property_Area'].isna().sum()
```

```python
df['Property_Area'].shape
```

```python
df1=pd.get_dummies(df['Property_Area'],prefix='Property_Area',drop_first=True)
df=pd.concat([df,df1],axis=1).drop('Property_Area',axis=1)
```

```python
df["Gender"].fillna(df["Gender"].mode()[0], inplace=True)
```

```python
df["Married"].fillna(df["Married"].mode()[0], inplace=True)
```

```python
df["Dependents"].fillna('3+', inplace=True)
```

```python
df["Self_Employed"].fillna(df["Self_Employed"].mode()[0], inplace=True)
```

```python
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)
```

```python
df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mode()[0], inplace=True)
```

```python
df["Credit_History"].fillna(1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df2=pd.get_dummies(df['Gender'],prefix='Gender',drop_first=True)
df=pd.concat([df,df2],axis=1).drop('Gender',axis=1)
```

```python
df3=pd.get_dummies(df['Married'],prefix='Married',drop_first=True)
df=pd.concat([df,df3],axis=1).drop('Married',axis=1)
```

```python
df4=pd.get_dummies(df['Education'],prefix='Education',drop_first=True)
df=pd.concat([df,df4],axis=1).drop('Education',axis=1)
```

```python
df.head(5)
```

```python
df5=pd.get_dummies(df['Self_Employed'],prefix='Self_Employed',drop_first=True)
df=pd.concat([df,df5],axis=1).drop('Self_Employed',axis=1)
```

```python
df6=pd.get_dummies(df['Loan_Status (Approved)'],prefix='Loan_Status (Approved)',drop_first=True)
df=pd.concat([df,df6],axis=1).drop('Loan_Status (Approved)',axis=1)
```

```python
df7=pd.get_dummies(df['Dependents'],prefix='Dependents',drop_first=True)
df=pd.concat([df,df7],axis=1).drop('Dependents',axis=1)
```
