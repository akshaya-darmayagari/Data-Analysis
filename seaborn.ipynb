# Seaborn-Statistical Data Visualization

## 1) What is Seaborn?

Seaborn is a Python library that builds on the top of Matplotlib that makes the statistical data more easy and beautiful.

## Why we need it?

1) For creating plot like violin plot,box plot and pair plots.

2) Automatically handles data from Pandas dataframes.

## Where we use it?

1) Data Analysis and reporting.

2) Exploring relationships between variables.

3) Machine learning for visualizing datasets and results.








## 2)Importing Seaborn

```python
import seaborn as sns
```

```python
## 4. Histograms & Countplots
# Histogram
import matplotlib.pyplot as plt
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
 
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='pink')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```

```python
# 1.Compare Passenger Survival by Class and Gender
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
plt.figure(figsize=(6,4))
sns.barplot(x='class', y='survived', hue='sex', data=titanic)
plt.title('Passenger Survival Rate by Class and Gender')
plt.ylabel('Survival Rate')
plt.show()


```

```python
# 2.Distribution of Passenger Ages
sns.histplot(titanic['age'], bins=30, kde=True, color='teal')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#3. Fare Distribution by Class
sns.boxplot(x='class', y='fare', data=titanic, palette='Set2')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# 4. Relationship Between Age and Fare
sns.scatterplot(x='age', y='fare', hue='class', data=titanic, palette='coolwarm')
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#5. Correlation Heatmap use titanic data set
corr = titanic.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```
