### 6. Analysis with NumPy & Pandas

```python
# Average student age
avg_age = np.mean(df['Student Age'])
print("Average Student Age:", avg_age)

# Average weekly study hours
avg_study_hours = np.mean(df['Weekly study hours'])
print("Average Weekly Study Hours:", avg_study_hours)

# Average reading frequency
avg_reading_freq = np.mean(df['Reading frequency'])
print("Average Reading Frequency:", avg_reading_freq)

# Average attendance to classes
avg_attendance = np.mean(df['Attendance to classes'])
print("Average Attendance to Classes:", avg_attendance)

# Average GPA by gender
avg_gpa_by_gender = df.groupby('Sex')['Cumulative grade point average in the last semester (/4.00)'].mean()
print("Average GPA by Gender:\n", avg_gpa_by_gender)
```

### Insights:
 
1.The average student age is 1.62 (encoded value), indicating most students fall into a younger age group.

2.On average, students study for about 2.2 hours per week, showing relatively low study duration.

3.The average reading frequency (1.94) suggests students engage in reading activities occasionally but not regularly.

4.Class attendance (1.24) is also low, meaning irregular participation in lectures.

5.Gender-based GPA analysis shows that students with code ‘2’ (likely female) have a higher average GPA (3.37) compared to ‘1’ (likely male) with 2.74, indicating better academic performance.

### 7. Visualization with Matplotlib

```python
import matplotlib.pyplot as plt

# Bar chart of GPA by gender
genders = avg_gpa_by_gender.index
gpa_values = avg_gpa_by_gender.values

plt.bar(genders, gpa_values, color='skyblue')
plt.title("Average GPA by Gender")
plt.xlabel("Gender")
plt.ylabel("Average GPA")
plt.show()
```

### Insights

1.The bar representing Gender 2 is visibly taller than Gender 1, showing that these students (likely females) achieve higher GPAs on average.

2.The GPA difference between the two genders is quite clear — Gender 2 averages around 3.38, while Gender 1 is around 2.74.

3.This gap suggests that Gender 2 students perform better academically across semesters.

4.The bar heights show consistent variation, implying the pattern is not random but trend-based.

5.Overall, the bar graph highlights a positive correlation between Gender 2 and higher academic performance.

```python
# Average GPA by Scholarship Type
avg_gpa_scholarship = df.groupby('Scholarship type')['Cumulative grade point average in the last semester (/4.00)'].mean()

plt.plot(avg_gpa_scholarship.index, avg_gpa_scholarship.values, marker='o', color='orange')
plt.title("Average GPA by Scholarship Type")
plt.xlabel("Scholarship Type")
plt.ylabel("Average GPA")
plt.show()
```

### Insights

1.The plot shows that some scholarship types are associated with higher GPAs, indicating better academic performance.

2.Other scholarship types have lower GPAs, suggesting less academic advantage or support.

3.Overall, there is a clear variation in GPA across scholarship types, highlighting the impact of scholarships on student performance.

### 8. Visualization with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of Student Age
sns.histplot(df['Student Age'], bins=15, kde=True)
plt.title("Distribution of Student Age")
plt.show()
```

### Insights

1.Most students are clustered around a younger age group, showing a concentrated age range in the dataset.

2.The distribution is slightly skewed, indicating that few students are older, but the majority are younger.

```python
df['GPA Category'] = df['Cumulative grade point average in the last semester (/4.00)'].apply(lambda x: 'Above 3.0' if x >= 3.0 else 'Below 3.0')

# Countplot of GPA Category by Gender
sns.countplot(x='GPA Category', hue='Sex', data=df)
plt.title("GPA Category Count by Gender")
plt.show()
```

### Insights

1.Gender 2 students (likely females) have a higher count in the “Above 3.0” GPA category, indicating better academic performance.

2.Gender 1 students (likely males) have more students in the “Below 3.0” GPA category, showing lower performance comparatively.

```python
sns.boxplot(x='Scholarship type', y='Cumulative grade point average in the last semester (/4.00)', data=df)
plt.title("GPA Distribution by Scholarship Type")
plt.show()
```

### Insights

1.The boxplot shows that some scholarship types have consistently higher GPAs, with smaller spread.

2.Other scholarship types show more variability in GPA, indicating differences in student performance within the same scholarship group.

```python
# Pairplot of numeric features with GPA Category as hue
sns.pairplot(df[['Student Age','Weekly study hours','Cumulative grade point average in the last semester (/4.00)','GPA Category']], hue='GPA Category')
plt.show()
```

### Insights

1.Students in the “Above 3.0” GPA category generally show higher study hours, indicating a positive correlation between study time and GPA.

2.The pairplot also shows that GPA distribution is slightly higher among younger students, though study hours seem more influential.

```python
# Compute correlation of numeric features
corr = df[['Student Age','Weekly study hours','Cumulative grade point average in the last semester (/4.00)']].corr()

# Heatmap of correlation
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Insights

1.Weekly Study Hours and GPA are likely positively correlated, showing that more study hours lead to higher academic performance.

2.Student Age shows little correlation with GPA, indicating age is not a strong factor in performance for this dataset.
